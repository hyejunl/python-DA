```python
import numpy as np
import pandas as pd

movie_df = pd.read_csv("./data/movies.csv")
rating_df = pd.read_csv("./data/ratings.csv")

#movie_df.columns #['movieId', 'title', 'genres']
# rating_df.columns #['userId', 'movieId', 'rating', 'timestamp']
```

```python
## 1. 사용자가 평가한 모든 영화의 전체 평균 평점

rating_df["rating"].mean()
```

```python
## 2. 각 사용자별 평균 평점


pd.DataFrame(rating_df["rating"].groupby(rating_df["userId"]).mean())

```

```python
## 3. 각 영화별 평균 평점

df2 = pd.merge(rating_df,movie_df, on="movieId", how="left")

pd.DataFrame(df2["rating"].groupby(df2["title"]).mean())


```

```python
## 4. 평균 평점이 가장 높은 영화의 제목 (동률이 있을 경우 모두 출력)

df2 = pd.merge(rating_df,movie_df, on="movieId", how="left")

df4 = pd.DataFrame(df2["rating"].groupby(df2["title"]).mean())

# df4.loc[df4["rating"]==df4["rating"].max()].index
df4.loc[df4["rating"]==df4["rating"].max()]
```
